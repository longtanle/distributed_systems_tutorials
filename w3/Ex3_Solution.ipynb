```python
import networkx as nx

G = nx.Graph()
G.add_edge('u','v',weight=2)
G.add_edge('u','w',weight=5)
G.add_edge('u','x',weight=1)
G.add_edge('v','x',weight=2)

G.add_edge('v','w',weight=3)
G.add_edge('x','w',weight=3)
G.add_edge('x','y',weight=1)
G.add_edge('y','w',weight=1)
G.add_edge('y','z',weight=2)
G.add_edge('w','z',weight=5)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
pos = nx.spring_layout(G)  # positions for all nodes
labels = nx.get_edge_attributes(G,'weight')
nx.draw(G,pos,labels={node:node for node in G.nodes()},edge_color='blue',node_color='red')
nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)
plt.show()
```

#### 2. Find the shortest paths using **shortest_path**

```python
print(nx.shortest_path(G, source='u', target='z', weight='weight'))
```

### 3. What is the total cost of the shortest path ?

```python
print(nx.shortest_path_length(G, source='u', target='z', weight='weight'))
```

### 4. What is the shortest path if all edge costs are equal to 1 ? 

```python
print(nx.shortest_path(G, source='u', target='z'))
```
